```python
top10_categories = (
    df.groupBy(col("Job Category"))
      .agg(count("*").alias("job_postings_count"))
      .orderBy(col("job_postings_count").desc())
      .limit(10)
)

top10_categories.show(truncate=False)
```

```python
# Step 1: Create average salary column
df = df.withColumn(
    "avg_salary_raw",
    (col("Salary Range From") + col("Salary Range To")) / 2
)

# Step 2: Normalize salary to Annual
df = df.withColumn(
    "annual_salary",
    when(col("Salary Frequency") == "Hourly", col("avg_salary_raw") * 2080)
    .when(col("Salary Frequency") == "Daily", col("avg_salary_raw") * 260)
    .otherwise(col("avg_salary_raw"))
)

# Step 3: Salary distribution per Job Category
salary_distribution = (
    df.groupBy("Job Category")
      .agg(
          avg("annual_salary").alias("avg_annual_salary")
      )
      .orderBy(col("avg_annual_salary").desc())
)

salary_distribution.show(truncate=False)
```

```python
df = df.withColumn(
    "degree_level",
    when(col("Minimum Qual Requirements").rlike("Doctor|PhD"), 4)
    .when(col("Minimum Qual Requirements").rlike("Master"), 3)
    .when(col("Minimum Qual Requirements").rlike("Bachelor"), 2)
    .when(col("Minimum Qual Requirements").rlike("Associate"), 1)
    .otherwise(0)
)

# --- Step 2: Remove null salaries ---
df_corr = df.filter(col("annual_salary").isNotNull())

# --- Step 3: Assemble features for correlation ---
assembler = VectorAssembler(
    inputCols=["degree_level", "annual_salary"],
    outputCol="features"
)

df_vector = assembler.transform(df_corr).select("features")

# --- Step 4: Compute Pearson Correlation ---
correlation_matrix = Correlation.corr(df_vector, "features").head()[0]

print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# --- Window partition by Agency ---
window_spec = Window.partitionBy("Agency").orderBy(col("annual_salary").desc())

# --- Rank salaries inside each agency ---
df_ranked = df.withColumn("rank", row_number().over(window_spec))

# --- Select highest salary job per agency ---
highest_salary_per_agency = (
    df_ranked
    .filter(col("rank") == 1)
    .select(
        "Agency",
        "Business Title",
        "annual_salary"
    )
    .orderBy(col("annual_salary").desc())
)

highest_salary_per_agency.show(truncate=False)
```

```python
# --- Step 1: Extract Year from Posting Date ---
df = df.withColumn("posting_year", year(col("Posting Date")))

# --- Step 2: Get latest year dynamically ---
latest_year = df.select(max("posting_year")).collect()[0][0]

# --- Step 3: Filter last 2 years dynamically ---
df_last_2_years = df.filter(col("posting_year") >= (latest_year - 1))

# --- Step 4: Calculate average annual salary per agency ---
avg_salary_per_agency = (
    df_last_2_years
    .groupBy("Agency")
    .agg(round(avg("annual_salary")).alias("avg_annual_salary"))
    .orderBy(col("avg_annual_salary").desc())
)

avg_salary_per_agency.show(truncate=False)
```

```python
from functools import reduce
from pyspark.sql import DataFrame


# Combine text
df = df.withColumn(
    "combined_text",
    lower(concat_ws(" ", col("Preferred Skills"), col("Job Description")))
)

skills = [
    "python","java","sql","scala","c++","r",
    "aws","azure","gcp","cloud","devops","kubernetes","docker",
    "spark","hadoop","kafka","etl","data warehouse",
    "machine learning","deep learning","data science",
    "power bi","tableau","excel","analytics",
    "mysql","postgresql","oracle","mongodb",
    "leadership","communication","project management"
]

results = []

for skill in skills:
    temp = (
        df.filter(col("combined_text").contains(skill))
          .agg(
              round(avg("annual_salary"), 0).alias("avg_salary"),
              count("*").alias("skill_count")
          )
          .withColumn("skill", lit(skill))
    )
    results.append(temp)


final_skills = reduce(DataFrame.unionByName, results)

# Remove very low occurrence skills (example: less than 5 postings)
final_skills = final_skills.filter(col("skill_count") >= 5)

# Order highest paid
final_skills.orderBy(col("avg_salary").desc()).select("skill","avg_salary").show(truncate=False)
```

```python
import matplotlib.pyplot as plt

# Convert Spark DataFrame to Pandas
top10_categories_pdf = top10_categories.toPandas()

# Sort ascending for horizontal bar clarity
top10_categories_pdf = top10_categories_pdf.sort_values(
    by="job_postings_count",
    ascending=True
)

plt.figure()

plt.barh(
    top10_categories_pdf["Job Category"],
    top10_categories_pdf["job_postings_count"]
)

plt.xlabel("Number of Job Postings")
plt.ylabel("Job Category")
plt.title("Top 10 Job Categories by Job Postings")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert Spark DataFrame to Pandas
salary_distribution_pdf = salary_distribution.toPandas()

# Take Top 10 highest paying categories for clarity
salary_distribution_pdf = salary_distribution_pdf.head(10)

# Sort ascending for better horizontal display
salary_distribution_pdf = salary_distribution_pdf.sort_values(
    by="avg_annual_salary",
    ascending=True
)

plt.figure()

plt.barh(
    salary_distribution_pdf["Job Category"],
    salary_distribution_pdf["avg_annual_salary"]
)

plt.xlabel("Average Annual Salary")
plt.ylabel("Job Category")
plt.title("Top 10 Job Categories by Average Annual Salary")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Compute average salary per degree level
degree_salary_avg = (
    df.select("degree_level", "annual_salary")
      .filter(col("annual_salary").isNotNull())
      .groupBy("degree_level")
      .avg("annual_salary")
      .orderBy("degree_level")
      .toPandas()
)

# Rename column for clarity
degree_salary_avg.rename(columns={
    "avg(annual_salary)": "avg_annual_salary"
}, inplace=True)

plt.figure()

plt.bar(
    degree_salary_avg["degree_level"],
    degree_salary_avg["avg_annual_salary"]
)

plt.xlabel("Degree Level (0=None, 1=Associate, 2=Bachelor, 3=Master, 4=Doctorate)")
plt.ylabel("Average Annual Salary")
plt.title("Average Salary by Education Level")

plt.xticks(degree_salary_avg["degree_level"])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert Spark DataFrame to Pandas
highest_salary_pdf = highest_salary_per_agency.toPandas()

# Take Top 10 highest paying agencies
highest_salary_pdf = highest_salary_pdf.head(10)

# Sort ascending for horizontal display
highest_salary_pdf = highest_salary_pdf.sort_values(
    by="annual_salary",
    ascending=True
)

plt.figure()

plt.barh(
    highest_salary_pdf["Agency"],
    highest_salary_pdf["annual_salary"]
)

plt.xlabel("Highest Annual Salary")
plt.ylabel("Agency")
plt.title("Top 10 Agencies by Highest Salary Posting")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert Spark DataFrame to Pandas
avg_salary_pdf = avg_salary_per_agency.toPandas()

# Take Top 10 agencies
avg_salary_pdf = avg_salary_pdf.head(10)

# Sort ascending for clean horizontal ranking
avg_salary_pdf = avg_salary_pdf.sort_values(
    by="avg_annual_salary",
    ascending=True
)

plt.figure(figsize=(10, 6))

bars = plt.barh(
    avg_salary_pdf["Agency"],
    avg_salary_pdf["avg_annual_salary"]
)

# Add value labels
for i, value in enumerate(avg_salary_pdf["avg_annual_salary"]):
    plt.text(value, i, f" {int(value):,}", va='center')

plt.xlabel("Average Annual Salary (Last 2 Years)")
plt.ylabel("Agency")
plt.title("Top 10 Agencies by Average Annual Salary (Last 2 Years)")

plt.grid(axis="x", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
