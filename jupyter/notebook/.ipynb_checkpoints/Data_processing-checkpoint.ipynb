```python
import findspark
findspark.init()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("pyspark-1") \
    .getOrCreate()
```

```python
df = spark.read.csv("/dataset/nyc-jobs.csv", header=True)
df.printSchema()
```

```python
df.columns
```

```python
def clean_data(df):
    
    # Remove duplicate job IDs
    df = df.dropDuplicates(["Job ID"])
    
    # Remove rows with null salary range
    df = df.filter(
        col("Salary Range From").isNotNull() &
        col("Salary Range To").isNotNull()
    )
    
    return df
```

```python
#Column Pre-Processing
def preprocess_columns(df):
    
    df = df.withColumn("Posting Date", to_date(col("Posting Date")))
    df = df.withColumn("Posting Updated", to_date(col("Posting Updated")))
    df = df.withColumn("Process Date", to_date(col("Process Date")))
    
    return df
```

```python
#Salary Normalization

def normalize_salary(df):
    
    df = df.withColumn(
        "avg_salary_raw",
        (col("Salary Range From") + col("Salary Range To")) / 2
    )
    
    df = df.withColumn(
        "annual_salary",
        when(col("Salary Frequency") == "Hourly", col("avg_salary_raw") * 2080)
        .when(col("Salary Frequency") == "Daily", col("avg_salary_raw") * 260)
        .otherwise(col("avg_salary_raw"))
    )
    
    return df
```

```python
#Data Wrangling

def wrangle_data(df):
    
    df = df.withColumn("posting_year", year(col("Posting Date")))
    
    # Remove unrealistic salary outliers
    df = df.filter(col("annual_salary") < 500000)
    
    return df
```

```python
#Text Transformation
def text_processing(df):
    
    df = df.withColumn(
        "job_desc_clean",
        lower(col("Job Description"))
    )
    
    return df
```

```python
#Full Processing Pipeline

def process_pipeline(df):
    
    df = clean_data(df)
    df = preprocess_columns(df)
    df = normalize_salary(df)
    df = wrangle_data(df)
    df = text_processing(df)
    
    return df
```

```python
df_processed = process_pipeline(df)

df_processed.printSchema()
df_processed.show(5)
```

```python

#Salary Binning (Salary Band Classification)
df_processed = df_processed.withColumn(
    "salary_band",
    when(col("annual_salary") < 50000, "Low")
    .when((col("annual_salary") >= 50000) & (col("annual_salary") < 100000), "Medium")
    .when((col("annual_salary") >= 100000) & (col("annual_salary") < 150000), "High")
    .otherwise("Very High")
)
```

```python

#Education Level Encoding (Ordinal Mapping)

df_processed = df_processed.withColumn(
    "education_level",
    when(lower(col("Minimum Qual Requirements")).rlike("doctor|phd"), 4)
    .when(lower(col("Minimum Qual Requirements")).rlike("master"), 3)
    .when(lower(col("Minimum Qual Requirements")).rlike("baccalaureate|bachelor"), 2)
    .when(lower(col("Minimum Qual Requirements")).rlike("associate"), 1)
    .otherwise(0)
)
```

```python

#Skill Count Feature

skills = [
    "python","java","sql","scala","c++","r",
    "aws","azure","gcp","cloud","devops","kubernetes","docker",
    "spark","hadoop","kafka","etl","data warehouse",
    "machine learning","deep learning","data science",
    "power bi","tableau","excel","analytics",
    "mysql","postgresql","oracle","mongodb",
    "leadership","communication","project management"
]

df_processed = df_processed.withColumn(
    "job_desc_lower",
    lower(col("Job Description"))
)

# Create skill array column
skill_array_expr = array()

for skill in skills:
    skill_array_expr = array_union(
        skill_array_expr,
        when(col("job_desc_lower").contains(skill), array(lit(skill)))
        .otherwise(array())
    )

df_processed = df_processed.withColumn("skills_found", skill_array_expr)

# Skill count from array size
df_processed = df_processed.withColumn(
    "skill_count",
    size(col("skills_found"))
)
df_processed.select(
    "annual_salary",
    "salary_band",
    "education_level",
    "skill_count",
    "skills_found"
).show( truncate=False)
```

```python
#Features removal based on the exploration/ profiling.
# ============================================
# FINAL FEATURE REMOVAL
# ============================================

columns_to_drop = [
    # Identifiers / High Cardinality
    "Job ID",
    "Business Title",
    "Title Code No",
    "Work Location 1",
    
    # Raw Salary Columns (replaced by annual_salary)
    "Salary Range From",
    "Salary Range To",
    "Salary Frequency",
    "avg_salary_raw",
    
    # Raw Text Columns (already engineered)
    "Job Description",
    "Minimum Qual Requirements",
    "Preferred Skills",
    "Additional Information",
    "To Apply",
    
    # Contact / Application Fields
    "Recruitment Contact",
    
    # Intermediate Columns
    "job_desc_lower"
]

df_final = df_processed.drop(*columns_to_drop)

df_final.printSchema()
df_final.show(5, truncate=False)
```

```python
df_final.write.mode("overwrite").parquet("/mnt/data/nyc_jobs_processed.parquet")
```

```python
# Convert array column to string (CSV doesn't support array type)
df_export = df_final.withColumn(
    "skills_found",
    concat_ws(", ", col("skills_found"))
)

# Save as CSV
df_export.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/mnt/data/processed_csv")
```
